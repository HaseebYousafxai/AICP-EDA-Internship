### **Task 1:**
> Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis
and a title:

```python
# Data for plotting the line
x = [0, 2, 4]
y = [1, 3, 0]

#set dataframe
data = {"x-axis": x, "y-axis": y}
df = pd.DataFrame(data)

# plot  data using matplotlib library
plt.plot(df["x-axis"], df["y-axis"])

# add title and labels to the axis
plt.xlabel('X plane')
plt.ylabel('Y plane')
plt.title('Line Plot')

plt.show() # line  will show after this command
df # dataframe  will be displayed in output
```

### **Task 2:**
> Write a Python program to plot two or more lines with legends, different widths and colors:

```python
x = [0, 20, 30]
y = [20, 40, 0]
y1 = [40, 0, 30]

plt.plot(x, y, label= 'Line 1', color = 'Blue', linewidth = 2.0)
plt.plot(x , y1, label = 'Line 2', color = 'Red', linewidth = 3.0, linestyle = '--')

plt.xlabel('X plane')
plt.ylabel('Y plane')
plt.title('Two lines of different colors and styles')


plt.legend()

plt.show()
```

> **Explanation**
- legend provides information about the elements or components represented in a chart, 
- helping users understand the meaning of colors, shapes, or lines used in the graphical representation.

### **Task 3:**
> Write a Python program to plot two or more lines and set the line markers:

```python
x = [0, 20, 30]
y = [20, 40, 0]
y1 = [40, 0, 30]

plt.plot(x, y, label= 'Line 1', marker='o', color = 'Blue', linewidth = 2.0, linestyle = 'dashdot')
plt.plot(x , y1, label = 'Line 2', marker='s', color = 'Red', linewidth = 3.0, linestyle = 'dotted')

plt.xlabel('X plane')
plt.ylabel('Y plane')
plt.title('Two lines of different colors and styles with marker')
plt.legend()

plt.show()
```

### **Task 5:**
> Write a Python program to create bar plot from a DataFrame. Sample Data Frame:
- x  a  b  c  d  e
- 2  4, 8, 5, 7, 6
- 4  2, 3, 4, 2, 6
- 6  4, 7, 4, 7, 8
- 8  2, 6, 4, 8, 6
- 10 2, 4, 3, 3, 2

```python

# create dataframe
data = {'a': [4, 2, 4, 2, 2],
        'b': [8, 3, 7, 6, 4],
        'c': [5, 4, 4, 4, 3],
        'd': [7, 2, 7, 8, 3],
        'e': [6, 6, 8, 6, 2]}

df = pd.DataFrame(data)
#bar plot
ax = df.plot(kind='bar', colormap='viridis', width=0.8)
#customize x plane values
custom_x_values = [2, 4, 6, 8, 10]
ax.set_xticks(range(len(custom_x_values)))
ax.set_xticklabels(custom_x_values) 

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot for Each Category with Custom X-axis Values')

plt.show()
```

### **Task 6:**
> Write a Python program to create a pie chart of gold medal achievements of five most successful
countries in 2016 Summer Olympics. Read the data from a csv file.
- Sample data:
- country,gold_medal
- United States,46
- Great Britain,27
- China,26
- Russia,19
- Germany,17

```python
# import csv file
df = pd.read_csv("olympics_data.csv")

# set distance between each category
explode_values = (0.06, 0, 0, 0, 0)

plt.pie(df["gold_medal"], labels=df["country"], autopct='%1.1f%%', startangle=140,
        colors=['gold', 'silver', 'brown', 'grey', 'orange'], explode=explode_values, shadow= True)


plt.title('Gold Medal Achievements in 2016 Summer Olympics')
plt.show()
```

> **explanation** 
- our dataset contain 2 columns: 1. country which contain country names. 
- 2. gold_medals  which contains the number of gold medals won by that country in respective categories.
- df["gold_medal"] represents  the column "gold_medals" and is a list containing all values under this column. and also for country coulmn.

### **Task 7:**
> Write a Python program to draw a scatter plot comparing two subject marks of Mathematics and
Science. Use marks of 10 students.
- Sample data:
- math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
- science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]
